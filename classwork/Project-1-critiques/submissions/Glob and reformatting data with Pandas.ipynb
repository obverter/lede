```python
import glob
```

```python
# find all filenames that end in html
filenames = glob.glob('*.html')
filenames
```

```python
filename = 'buerglererich_619403_link.html'
open(filename).read()
```

I want two things:
1. A list of filenames (we have that)
2. A list of file contents

Basically means I want to convert each filename into its contents
**Use a list comprehension**

```python
contents = [open(filename).read() for filename in filenames]
```

```python
import pandas as pd
```

```python
# Hey pd.DataFrame, here's a list for each column
df = pd.DataFrame({
    'filename': filenames,
    'content': contents
})
df
```

```python
# Easy-peasy extracting of things with .str.extract

# STEP ONE: Copy in an example of what you want
# STEP TWO: Wrap what you're interested in using parens
# STEP THREE: Replace the stuff inside the parents with .*
df['name'] = df.content.str.extract("<title>Project 1: Your very first project.: (.*)</title>")
```

```python
# Now do this for the URL
# <a href="https://petebrown.github.io/project-01/">Click Here to go to the submission</a>

df.content.str.extract('<a href="https://(.*).github.io/(.*)">Click Here to go to the submission</a>')
```
